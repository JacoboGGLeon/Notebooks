```python
def address_to_location(address, key):
    import requests

    GOOGLE_MAPS_API_URL = 'https://maps.googleapis.com/maps/api/geocode/json'

    params = {
        'address': address, #'calle Yuriria, de la manzana 26, lote 12, en la colonia Valle de Anáhuac',
        'sensor': 'false',
        'key': key #'AIzaSyAELcgNJs68gefdSp5TUxk73t8IUh3vTME'
    }

    # Do the request and get the response data
    req = requests.get(GOOGLE_MAPS_API_URL, params=params)
    res = req.json()

    # Use the first result
    result = res['results'][0]

    geodata = dict()
    geodata['lat'] = result['geometry']['location']['lat']
    geodata['lng'] = result['geometry']['location']['lng']
    geodata['address'] = result['formatted_address']
    
    #res['results'][0]['geometry']['location']
    #r = geodata['address']

    #print('{address}. (lat, lng) = ({lat}, {lng})'.format(**geodata))
    # 221B Baker Street, London, Greater London NW1 6XE, UK. (lat, lng) = (51.5237038, -0.1585531)
    
    return (geodata['address'], {'lat':geodata['lat'], 'lng': geodata['lng']})
```

```python
def location_to_nearby_places(location, radius, key):        
    import requests
    
    GOOGLE_MAPS_API_URL = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

    params = {
        'location': location,#'19.5179648, -99.0273843', #'19.5179648, -99.0273843', #'calle Yuriria, de la manzana 26, lote 12, en la colonia Valle de Anáhuac',
        'radius': radius,
        'key': key
    }

    # Do the request and get the response data
    req = requests.get(GOOGLE_MAPS_API_URL, params=params)
    res = req.json()

    df = pd.DataFrame(res['results'])
    df_copy = df.copy()
    df_copy.reset_index(inplace=True)

    df_geometry = pd.DataFrame(columns=['index', 'location', 'lat', 'lng'])

    for line in range(len(df_copy['geometry'])):
        location = df_copy['geometry'][line]
        df_geometry = df_geometry.append({'index': line,
                                          'location': location.get('location'),
                                          'lat': location.get('location').get('lat'), 
                                          'lng': location.get('location').get('lng')}, ignore_index=True)

    df_selected= pd.DataFrame()
    df_selected = pd.merge(left=df_copy, right=df_geometry, how='outer', )
    df_selected = df_selected[['name', 'types', 'vicinity', 'location', 'lat', 'lng']]

    df_selected

    return df_selected
```
