# Carga de Datos

## 2014

```python
path_sdemt114 = '../../Datasets/ENOE/2014/2014trim1_csv/sdemt114.CSV'

df_sdemt114 = pd.read_csv(path_sdemt114, low_memory=False, encoding='latin')

df_sdemt114.head()
```

```python
path_sdemt214 = '../../Datasets/ENOE/2014/2014trim2_csv/sdemt214.CSV'

df_sdemt214 = pd.read_csv(path_sdemt214, low_memory=False, encoding='latin')

df_sdemt214.head()
```

```python
path_sdemt314 = '../../Datasets/ENOE/2014/2014trim3_csv/sdemt314.CSV'

df_sdemt314 = pd.read_csv(path_sdemt314, low_memory=False, encoding='latin')

df_sdemt314.head()
```

```python
path_sdemt414 = '../../Datasets/ENOE/2014/2014trim4_csv/sdemt414.CSV'

df_sdemt414 = pd.read_csv(path_sdemt414, low_memory=False, encoding='latin')

df_sdemt414.head()
```

Se copia la información para no volver a cargar todo

```python
df_copy_sdemt114 = df_sdemt114.copy()
df_copy_sdemt214 = df_sdemt214.copy()
df_copy_sdemt314 = df_sdemt314.copy()
df_copy_sdemt414 = df_sdemt414.copy()
```

```python
print('sdemt114: ', df_copy_sdemt114.shape)
print('sdemt214: ', df_copy_sdemt214.shape)
print('sdemt314: ', df_copy_sdemt314.shape)
print('sdemt414: ', df_copy_sdemt414.shape)
print('total: ', 
      df_copy_sdemt114.shape[0]+
      df_copy_sdemt214.shape[0]+
      df_copy_sdemt314.shape[0]+
      df_copy_sdemt414.shape[0])
```

```python
#concatenar
frames = [df_copy_sdemt114, 
          df_copy_sdemt214, 
          df_copy_sdemt314, 
          df_copy_sdemt414]

df_sdemt_2014 = pd.concat(frames)
df_sdemt_2014.shape
```

## Exportar a CSV

```python
#exportar csv

path_df_sdemt_2014 = '../../Datasets/ENOE/dataframes/sdemt_2014.csv'

df_sdemt_2014.to_csv(path_or_buf=path_df_sdemt_2014, index=True, encoding='utf-8')
```
