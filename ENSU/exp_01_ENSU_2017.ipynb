```python
ensu_viv_117 = '../../Datasets/ENSU/data/ensu_bd_marzo_2017_dbf/ENSU_VIV_0317.csv'

df_ensu_viv_117 = pd.read_csv(ensu_viv_117, low_memory=False, index_col=0)

df_ensu_viv_117.head()
```

```python
ensu_cs_117 = '../../Datasets/ENSU/data/ensu_bd_marzo_2017_dbf/ENSU_CS_0317.csv'

df_ensu_cs_117 = pd.read_csv(ensu_cs_117, low_memory=False, index_col=0)

df_ensu_cs_117.head()
```

```python
ensu_cb_117 = '../../Datasets/ENSU/data/ensu_bd_marzo_2017_dbf/ENSU_CB_0317.csv'

df_ensu_cb_117 = pd.read_csv(ensu_cb_117, low_memory=False, index_col=0)

df_ensu_cb_117.head()
```

### ConstrucciÃ³n de IDs

```python
id_ensu_viv_117 = (df_ensu_viv_117['CVE_ENT'] +
                   df_ensu_viv_117['UPM'] +
                   df_ensu_viv_117['VIV_SEL'])

id_ensu_viv_117.count()
```

```python
id_ensu_cs_117 = (df_ensu_cs_117['CVE_ENT'] +
                  df_ensu_cs_117['UPM'] +
                  df_ensu_cs_117['VIV_SEL'] +
                  df_ensu_cs_117['N_REN'])

id_ensu_cs_117.count()
```

```python
id_ensu_cb_117 = (df_ensu_cb_117['CVE_ENT'] +
                   df_ensu_cb_117['UPM'] +
                   df_ensu_cb_117['VIV_SEL'] +
                   df_ensu_cb_117['R_SEL'])

id_ensu_cb_117.count()
```

```python
#copiamos para proteger la info
df_copy_ensu_viv_117 = df_ensu_viv_117.copy()
df_copy_ensu_cs_117 = df_ensu_cs_117.copy()
df_copy_ensu_cb_117 = df_ensu_cb_117.copy()
```

```python
print(df_copy_ensu_viv_117.shape)
print(df_copy_ensu_cs_117.shape)
print(df_copy_ensu_cb_117.shape)
```

```python
#se agrega la columna ID
df_copy_ensu_viv_117['ID'] = id_ensu_viv_117

#se coloca el ID previamente calculado
df_copy_ensu_viv_117.set_index(keys='ID', inplace=True)
df_copy_ensu_viv_117.head()
```

```python
#se agrega la columna ID
df_copy_ensu_cs_117['ID'] = id_ensu_cs_117

#se coloca el ID previamente calculado
df_copy_ensu_cs_117.set_index(keys='ID', inplace=True)
df_copy_ensu_cs_117.head()
```

```python
#se agrega la columna ID
df_copy_ensu_cb_117['ID'] = id_ensu_cb_117

#se coloca el ID previamente calculado
df_copy_ensu_cb_117.set_index(keys='ID', inplace=True)
df_copy_ensu_cb_117.head()
```

```python
#exportar csv

path = '../../Datasets/ENSU/dataframes/seguridad_urbana_117.csv'

df_copy_ensu_cb_117.to_csv(path_or_buf=path, index=True, encoding='utf-8')
```
