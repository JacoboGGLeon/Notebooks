```python
list_CVE_ENT = df_exp_1_ENSU_CB_2017.CVE_ENT.astype('category').cat.categories.tolist()
list_CVE_ENT
```

```python
secc_01_list_CVE_ENT = list_CVE_ENT[:32]
secc_02_list_CVE_ENT = list_CVE_ENT[32:64]
secc_03_list_CVE_ENT = list_CVE_ENT[64:]

df_list_CVE_ENT = pd.DataFrame({'secc_01':secc_01_list_CVE_ENT,
                                'secc_02':secc_02_list_CVE_ENT,
                                'secc_03':secc_03_list_CVE_ENT})

catalogo_CVE_ENT_01 = {df_list_CVE_ENT.loc[idx, 'secc_01']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}
catalogo_CVE_ENT_02 = {df_list_CVE_ENT.loc[idx, 'secc_02']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}
catalogo_CVE_ENT_03 = {df_list_CVE_ENT.loc[idx, 'secc_03']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}

catalogo_CVE_ENT = {}
catalogo_CVE_ENT.update(catalogo_CVE_ENT_01)
catalogo_CVE_ENT.update(catalogo_CVE_ENT_02)
catalogo_CVE_ENT.update(catalogo_CVE_ENT_03)
catalogo_CVE_ENT
```

```python
df_exp_1_ENSU_CB_2017.CVE_ENT = df_exp_1_ENSU_CB_2017.CVE_ENT.map(catalogo_CVE_ENT)
df_exp_1_ENSU_CB_2017.CVE_ENT.head()
```
