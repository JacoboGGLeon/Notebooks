```python
ensu_cb_115 = '../../Datasets/ENSU/data/ensu_bd_2015_dbf/ensu_bd_mar_2015_dbf/ENSU_CB_0315.csv'

df_ensu_cb_115 = pd.read_csv(ensu_cb_115, low_memory=False, index_col=0)

df_ensu_cb_115.head()
```
