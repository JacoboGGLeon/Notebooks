```python
list_CVE_ENT = df_exp_1_ENSU_CB_2017.CVE_ENT.astype('category').cat.categories.tolist()
list_CVE_ENT
```

```python
secc_01_list_CVE_ENT = list_CVE_ENT[:32]
secc_02_list_CVE_ENT = list_CVE_ENT[32:64]
secc_03_list_CVE_ENT = list_CVE_ENT[64:]

df_list_CVE_ENT = pd.DataFrame({'secc_01':secc_01_list_CVE_ENT,
                                'secc_02':secc_02_list_CVE_ENT,
                                'secc_03':secc_03_list_CVE_ENT})

catalogo_CVE_ENT_01 = {df_list_CVE_ENT.loc[idx, 'secc_01']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}
catalogo_CVE_ENT_02 = {df_list_CVE_ENT.loc[idx, 'secc_02']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}
catalogo_CVE_ENT_03 = {df_list_CVE_ENT.loc[idx, 'secc_03']: df_list_CVE_ENT.loc[idx, 'secc_03'] for idx in range(df_list_CVE_ENT.shape[0])}

catalogo_CVE_ENT = {}
catalogo_CVE_ENT.update(catalogo_CVE_ENT_01)
catalogo_CVE_ENT.update(catalogo_CVE_ENT_02)
catalogo_CVE_ENT.update(catalogo_CVE_ENT_03)
catalogo_CVE_ENT
```

```python
df_exp_1_ENSU_CB_2017.CVE_ENT = df_exp_1_ENSU_CB_2017.CVE_ENT.map(catalogo_CVE_ENT)
df_exp_1_ENSU_CB_2017.CVE_ENT.head()
```

### INEGI 
## Catálogo de Claves de Entidades Federativas, Municipios y Localidades

[referencia](http://www.inegi.org.mx/geo/contenidos/geoestadistica/catalogoclaves.aspx)

```python
CAT_INEGI_2018 = '../../Datasets/INEGI_LOC.csv'

df_CAT_INEGI_2018 = pd.read_csv(CAT_INEGI_2018, low_memory=False, index_col=0, encoding='latin_1')

df_CAT_INEGI_2018.reset_index().head()
```

```python
df_CAT_INEGI_2018.shape
```

+ `CVE_LOC`: CLAVE DE LOCALIDAD 
+ `NOM_LOC`: NOMBRE DE LOCALIDAD
+ `POB_TOTAL`: POBLACIÓN TOTAL
+ `POB_MAS`: POBLACIÓN MASCULINA
+ `POB_FEM`: POBLACIÓN FEMENINA
+ `TOT_VIV`: TOTAL DE VIVIENDAS HABITADAS

```python
df_CAT_INEGI_2018.rename(columns={'Z1':'POB_TOTAL', 'Z2':'POB_MAS', 'Z3':'POB_FEM', 'Z119':'TOT_VIV'}, inplace=True)
df_CAT_INEGI_2018.reset_index(inplace=True)
df_CAT_INEGI_2018.head()
```

```python
df_CAT_INEGI_2018.CVE_LOC.groupby(df_CAT_INEGI_2018.CVE_LOC).count()
```

```python
df_CAT_INEGI_2018.drop(df_CAT_INEGI_2018[(df_CAT_INEGI_2018.POB_TOTAL == '-') | (df_CAT_INEGI_2018.POB_FEM == '*') | (df_CAT_INEGI_2018.POB_MAS == '*')].index, inplace=True)
```

```python
df_CAT_INEGI_2018[(df_CAT_INEGI_2018.POB_TOTAL == '-') | (df_CAT_INEGI_2018.POB_FEM == '*') | (df_CAT_INEGI_2018.POB_MAS == '*')]
```

```python
df_CAT_INEGI_2018_POB_TOTAL = pd.DataFrame(df_CAT_INEGI_2018.POB_TOTAL.astype('int64').groupby(df_CAT_INEGI_2018.CVE_LOC).sum())
df_CAT_INEGI_2018_POB_MAS = pd.DataFrame(df_CAT_INEGI_2018.POB_MAS.astype('int64').groupby(df_CAT_INEGI_2018.CVE_LOC).sum())
df_CAT_INEGI_2018_POB_FEM = pd.DataFrame(df_CAT_INEGI_2018.POB_FEM.astype('int64').groupby(df_CAT_INEGI_2018.CVE_LOC).sum())

df_01 = pd.merge(df_CAT_INEGI_2018_POB_MAS.reset_index(), df_CAT_INEGI_2018_POB_FEM.reset_index(), how='inner', on='CVE_LOC')
df = pd.merge(df_01.reset_index(), df_CAT_INEGI_2018_POB_TOTAL.reset_index(), how='inner', on='CVE_LOC')

df.head()
```
