```python
import pandas as pd
from simpledbf import Dbf5

# convert from DBF to CSV

in_dbf = '../Datasets/ENSU/data/ensu_bd_dic_2017_dbf/ENSU_VIV_1217.dbf'
out_csv = '../Datasets/ENSU/data/ensu_bd_dic_2017_dbf/ENSU_VIV_1217.csv'

dbf = Dbf5(in_dbf, codec='latin_1')
dbf.to_csv(out_csv)

df = pd.read_csv(out_csv, encoding='latin-1', low_memory=False)
df.to_csv(out_csv, encoding='utf-8')

df.head()
```
