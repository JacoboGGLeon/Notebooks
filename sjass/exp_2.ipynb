```python
def check_value_in_list(list_for_check, value_for_check):
    result = False
    
    # traverse in the list
    for value_temporal in list_for_check:
        # compare with all the values
        # with val
#        print(value_temporal)
        if value_for_check == value_temporal:
#            print(value_temporal)
            result = True
    return result
```

```python
def address_components_to_dict(resource, dictionary):
    for line in resource:
    #    print(line['types'])
    #    print(check_value_in_list(line['types'], 'locality'))

        #find street_number
        if(check_value_in_list(line['types'], 'street_number') == True):
#            print('street_number: ' + line['long_name'])
            dictionary['street_number'] = line['long_name']
        else:
            dictionary['street_number'] = 'NaN'
            
        #find route
        if(check_value_in_list(line['types'], 'route') == True):
#            print('route: ' + line['long_name'])
            dictionary['route'] = line['long_name']
        else:
            dictionary['route'] = 'NaN'

        #find sublocality
        if(check_value_in_list(line['types'], 'sublocality') == True):        
#            print('sublocality: ' + line['long_name'])
            dictionary['sublocality'] = line['long_name']
        else:
            dictionary['sublocality'] = 'NaN'

        #find locality
        if(check_value_in_list(line['types'], 'locality') == True):        
#            print('locality: ' + line['long_name'])
            dictionary['locality'] = line['long_name']
        else:
            dictionary['locality'] = 'NaN'

        #find administrative_area
        if(check_value_in_list(line['types'], 'administrative_area_level_1') == True):        
#            print('administrative_area: ' + line['long_name'])
            dictionary['administrative_area'] = line['long_name']
        else:
            dictionary['administrative_area'] = 'NaN'

        #find country
        if(check_value_in_list(line['types'], 'country') == True):        
#            print('country: ' + line['long_name'])
            dictionary['country'] = line['long_name']
        else:
            dictionary['country'] = 'NaN'
    return dictionary
```
